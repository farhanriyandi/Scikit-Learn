# Bayes' Theorem

Bayes' Theorem menawarkan suatu formula untuk menghitung nilai probability dari suatu event dengan memanfaatkan pengetahuan sebelumnya dari kondisi terkait; atau sering kali dikenal dengan istilah conditional probability.

# Dataset: Breast Cancer Wisconsin (Diagnostic)

# Load Dataset

```python
from sklearn.datasets import load_breast_cancer

print(load_breast_cancer().DESCR)
```

```python
#load_breast_cancer?
X, y = load_breast_cancer(return_X_y=True)
X.shape
```

# Training & Testing set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size=0.2,
                                                   random_state=0)
#Random state ini istilah lain untuk random seed number. Ini biasa digunakan ketika kita membutuhkan mekanisme pengacakan (randomization) yang replicable atau bisa diulang dengan hasil yang konsisten

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

# Naive Bayes dengan Scikit Learn

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

model = GaussianNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
model.score(X_test, y_test)
```
